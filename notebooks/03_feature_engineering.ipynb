```python
import pandas as pd

X_train = pd.read_csv('../data/processed/X_train.csv')
X_val = pd.read_csv('../data/processed/X_val.csv')
X_test = pd.read_csv('../data/processed/X_test.csv')

y_train = pd.read_csv('../data/processed/y_train.csv').squeeze()
y_val = pd.read_csv('../data/processed/y_val.csv').squeeze()
y_test = pd.read_csv('../data/processed/y_test.csv').squeeze()
```

```python
late_cols = ['NumberOfTime30-59DaysPastDueNotWorse',
             'NumberOfTime60-89DaysPastDueNotWorse',
             'NumberOfTimes90DaysLate']

X_train['TotalTimesLate'] = X_train[late_cols].sum(axis=1)
X_val['TotalTimesLate']   = X_val[late_cols].sum(axis=1)
X_test['TotalTimesLate']  = X_test[late_cols].sum(axis=1)

X_train[['NumberOfTime30-59DaysPastDueNotWorse',
         'NumberOfTime60-89DaysPastDueNotWorse',
         'NumberOfTimes90DaysLate', 'TotalTimesLate']].head()
```

```python
X_train['TotalTimesLate'].describe()
```

```python
# Does this new feature separate defaulters from non-defaulters?
combined = pd.concat([X_train['TotalTimesLate'], y_train], axis=1)
combined.groupby('SeriousDlqin2yrs')['TotalTimesLate'].mean()
```

```python
X_train['IncomePerDependent'] = X_train['MonthlyIncome'] / (X_train['NumberOfDependents'] + 1)
X_val['IncomePerDependent']   = X_val['MonthlyIncome'] / (X_val['NumberOfDependents'] + 1)
X_test['IncomePerDependent']  = X_test['MonthlyIncome'] / (X_test['NumberOfDependents'] + 1)

X_train[['MonthlyIncome', 'NumberOfDependents', 'IncomePerDependent']].head()
```

```python
X_train['IncomePerDependent'].describe()
```

```python
combined = pd.concat([X_train['IncomePerDependent'], y_train], axis=1)
combined.groupby('SeriousDlqin2yrs')['IncomePerDependent'].mean()
```

```python
X_train['AgeGroup'] = pd.cut(X_train['age'], bins=[0, 30, 45, 60, 110], 
                                labels=['<30', '30-45', '45-60', '60+'])
X_val['AgeGroup']   = pd.cut(X_val['age'], bins=[0, 30, 45, 60, 110], 
                                labels=['<30', '30-45', '45-60', '60+'])
X_test['AgeGroup']  = pd.cut(X_test['age'], bins=[0, 30, 45, 60, 110], 
                                labels=['<30', '30-45', '45-60', '60+'])

X_train['AgeGroup'].value_counts()
```

```python
combined = pd.concat([X_train['AgeGroup'], y_train], axis=1)
combined.groupby('AgeGroup', observed=True)['SeriousDlqin2yrs'].mean()
```

```python
# One-hot encode the AgeGroup feature
X_train = pd.get_dummies(X_train, columns=['AgeGroup'], drop_first=True)
X_val   = pd.get_dummies(X_val, columns=['AgeGroup'], drop_first=True)
X_test  = pd.get_dummies(X_test, columns=['AgeGroup'], drop_first=True)
```

```python
# Save the feature engineered datasets
X_train.to_csv('../data/processed/X_train_fe.csv', index=False)
X_val.to_csv('../data/processed/X_val_fe.csv', index=False)
X_test.to_csv('../data/processed/X_test_fe.csv', index=False)
```
